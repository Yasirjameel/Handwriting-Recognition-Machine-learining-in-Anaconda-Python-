```python
#yasir jamil
#bscs 8th
#2591
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train = np.load('x_train.npy')
y_train = np.load('y_train.npy')
X_test = np.load('x_val.npy')
y_test = np.load('y_val.npy')
```

```python
fig = plt.figure(figsize=(8, 8))
n_cols = 2 
n_rows = 2
for i in range(1, n_rows*n_cols +1):
    img = X_train[i, :, :]
    fig.add_subplot(n_rows, n_cols, i)
    plt.imshow(img, cmap=plt.cm.gray)
plt.show()
```

```python
model = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=10, min_samples_leaf=3)
model.fit(X_train.reshape(-1, 28*28), y_train)
```

```python
print(f'train accuracy = { accuracy_score(model.predict(X_train.reshape(-1,28*28)), y_train)}')
print(f'test accuracy = { accuracy_score(model.predict(X_test.reshape(-1,28*28)), y_test)}')
```

```python
x_test_reshaped = X_test.reshape(-1,28*28)
predictions = model.predict(x_test_reshaped)
misclassified =  x_test_reshaped[np.where((predictions) !=  y_test) ]
```

```python
random_indicies = np.random.randint(0,200, 20)
fig = plt.figure(figsize=(20,10))
n_cols = 5
n_rows = 2
for i in range(1, n_rows*n_cols +1):
    img = misclassified[random_indicies[i-1], :].reshape(28,28)
    fig.add_subplot(n_rows, n_cols, i)
    plt.imshow(img, cmap=plt.cm.gray)
    plt.title(f'label = {y_test[random_indicies[i-1]]}', fontsize = 20)
plt.show()
```
